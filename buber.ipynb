```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_city_ride = pd.merge(ride_df,city_df)
#print(city_df.head())
#print(ride_df.head())

# Display the data table for preview
merge_city_ride
```

```python
city_df.head()
```

```python
city_name_list = merge_city_ride['city'].unique()
city_type_list = merge_city_ride['type'].unique()
    

groupby_city = merge_city_ride.groupby('city').sum()
groupby_city['number of rides'] = merge_city_ride['city'].value_counts().to_frame().sort_index()
groupby_city['average fare'] = round(groupby_city['fare'] / groupby_city['number of rides'],2)
groupby_city['driver_count'] = groupby_city['driver_count']/groupby_city['number of rides']
groupby_city['type'] = pd.Series(index=city_df.city.values, data=city_df.type.values)


#print(round(groupby_city.iloc[1,0]/groupby_city.iloc[1,3],2))

#avg_fare_per_city = []
#for i in np.arange(len(city_name_list)):
#    avg_fare_per_city[i] = (groupby_city.iloc[i,0]/groupby_city.iloc[i,3])

del groupby_city['ride_id']
groupby_city.rename(columns={'driver_count': 'driver count','fare': 'total fares'})

    
```

### Bubble Plot

```python
# Obtain the x and y coordinates for each of the three city types
total_number_of_rides = groupby_city['number of rides']
average_fare = groupby_city['average fare']
driver_count_per_city = groupby_city['driver_count']
# Build the scatter plots for each city types
city_types = groupby_city.type.unique()
print(city_types)

urban_cities = pd.DataFrame()
urban_cities = groupby_city[groupby_city.type==city_types[0]]
urban_cities_numb_rides = urban_cities['number of rides']
urban_cities_avg_fare = urban_cities['average fare']
urban_cities_driver_count = urban_cities['driver_count']


suburban_cities = pd.DataFrame()
suburban_cities = groupby_city[groupby_city.type==city_types[1]]
suburban_cities_numb_rides = suburban_cities['number of rides']
suburban_cities_avg_fare = suburban_cities['average fare']
suburban_cities_driver_count = suburban_cities['driver_count']

rural_cities = pd.DataFrame()
rural_cities = groupby_city[groupby_city.type==city_types[2]]
rural_cities_numb_rides = rural_cities['number of rides']
rural_cities_avg_fare = rural_cities['average fare']
rural_cities_driver_count = rural_cities['driver_count']


#for ctype in city_types:
#    cities = groupby_city[groupby_city.type==ctype]
#    print(cities)
    
#Incorporate the other graph properties
plt.grid()
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Pyber Ride Sharing Data (2016)')

plt.scatter(urban_cities_numb_rides,urban_cities_avg_fare, s=urban_cities_driver_count*10, alpha=.75)
plt.scatter(suburban_cities_numb_rides,suburban_cities_avg_fare, s=suburban_cities_driver_count*10, alpha=.75)
plt.scatter(rural_cities_numb_rides,rural_cities_avg_fare, s=rural_cities_driver_count*10, alpha=.75)

lgnd = plt.legend(('urban','suburban','rural'))

lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

#plt.scatter(total_number_of_rides,average_fare, s=driver_count_per_city,)
# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
groupby_city['fare'].sum()
```

### Total Fares by City Type

```python
total_fares = groupby_city['fare'].sum()

urban_cities_fares_percs = round(urban_cities['fare'].sum(),2)/total_fares
suburban_cities_fares_percs = round(suburban_cities['fare'].sum(),2)/total_fares
rural_cities_fares_percs = round(rural_cities['fare'].sum(),2)/total_fares

city_fares_percs = [urban_cities_fares_percs,suburban_cities_fares_percs,rural_cities_fares_percs]
explode = [.1,0,0]
labels = ['urban','suburban','rural']

plt.title('% of Total Fares by City Type')
plt.pie(city_fares_percs,labels=labels,explode=explode,autopct='%1.1f%%',startangle=-90,shadow=True)
```

### Total Rides by City Type

```python
perc_of_urban_rides = round(len(merge_city_ride[merge_city_ride['type']=='Urban'])/len(merge_city_ride)*100,1)
perc_of_suburban_rides = round(len(merge_city_ride[merge_city_ride['type']=='Suburban'])/len(merge_city_ride)*100,1)
perc_of_rural_rides = round(len(merge_city_ride[merge_city_ride['type']=='Rural'])/len(merge_city_ride)*100,1)

perc_of_type_rides = [perc_of_urban_rides,perc_of_suburban_rides,perc_of_rural_rides]
labels = ['urban','suburban','rural']
explode = [0.1, 0, 0]

plt.title('% of Total Rides by City Type')
plt.pie(perc_of_type_rides,explode=explode,labels=labels,autopct='%1.1f%%',startangle=-90,shadow=True)
```

### Total Drivers by City Type

```python
total_driver_count = groupby_city['driver_count'].sum()

urban_driver_count = urban_cities['driver_count'].sum()
urban_driver_count_perc = round(urban_driver_count/total_driver_count*100,1)

suburban_driver_count = suburban_cities['driver_count'].sum()
suburban_driver_count_perc = round(suburban_driver_count/total_driver_count*100,1)

rural_driver_count = rural_cities['driver_count'].sum()
rural_driver_count_perc = round(rural_driver_count/total_driver_count*100,1)

driver_count_percs = [urban_driver_count_perc,suburban_driver_count_perc,rural_driver_count_perc]
labels = ['urban','suburban','rural']
explode = [0.1,0,0]

plt.title('% of Total Drivers by City Count')
plt.pie(driver_count_percs,labels=labels,explode=explode,shadow=True,startangle=-135,autopct='%1.1f%%')
```
